```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
s2 = s + 1
fig, ax = plt.subplots()
ax.plot(t, s)
ax.plot(t, s2)
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()
ax.properties()['children'][1].set_color('green')
plt.show()
```
