```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
import re

# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
foldername = ''
save_name = 'trace6961'

n_begin = 2000000
n_final = 2001000
n_step = 1

filename = 'dump' + str(step) + '.atoms'

with open(filename) as f:
    
    lines = f.readlines()
    id_head_begin = [n for n, line in enumerate(lines) if line[0:11]=='ITEM: ATOMS'][0]
    header = lines[id_head_begin].split(' ')[2:-1]
    header = header
    
step = n_begin
filename = 'dump' + str(step) + '.atoms'
datapd = pd.read_csv(filename, delim_whitespace = True, skiprows = id_head_begin+1, header = None, names = header)

datapd = datapd.loc[datapd['id'].isin([6961,6775,6773,6748] ) ]

datapd.insert(loc = 0, column = 'step', value = step)
    
for step in range(n_begin+1, n_final+1, n_step):
    
    filename = 'dump' + str(step) + '.atoms'
    data_onestep = pd.read_csv(filename, delim_whitespace = True, skiprows = id_head_begin+1, header = None, names = header)
    
    data_onestep = data_onestep.loc[data_onestep['id'].isin([6961,6775,6773,6748] ) ]
    
    data_onestep.insert(loc = 0, column = 'step', value = step)
    
    
    #data_onestep = np.loadtxt(filename,skiprows=9)
    
    datapd = pd.concat([datapd, data_onestep], ignore_index = True)
    
datapd.to_csv(save_name, index = False)
```

```python
variable_groupby = 'id'
select_group = 6775

scale_x = 'linear'
scale_y = 'linear'

x_variable_name = 'step'
y1_variable_name = 'fx'
y2_variable_name = 'fy'
y3_variable_name = 'fz'
y_variable_name = y1_variable_name + y2_variable_name +y3_variable_name

timestep_start = 1
timestep_end = -1

datapd_select = datapd.loc[datapd[variable_groupby] == select_group]

x = datapd_select[x_variable_name].values
y1 = datapd_select[y1_variable_name].values
y2 = datapd_select[y2_variable_name].values
y3 = datapd_select[y3_variable_name].values
y = (y1**2+y2**2+y3**3)**0.5

fig = plt.figure()
plt.xscale(scale_x)
plt.yscale(scale_y)
plt.xlabel(x_variable_name)
plt.ylabel(y_variable_name)
plt.plot(x,y,'.')
fig.savefig(foldername + x_variable_name + '_vs_' + y_variable_name + '_step_' + str(timestep_start) + 'to' + str(timestep_end) + '_scale' +scale_x + scale_y + '.png')   # save the figure to file
```
