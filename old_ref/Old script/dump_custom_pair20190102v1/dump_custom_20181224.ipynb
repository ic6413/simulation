```python
import re
import time
from itertools import chain
from itertools import repeat
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

runfolder = '0101_mu05_freeze'
readpath = '/home/ic6413/mylammps/myfiles/run/' + runfolder + '/output/dump/'
filename = 'nb_maxKEatom_0.dump'
outputpath = '/home/ic6413/mylammps/myfiles/run/' + runfolder + '/output/python/'

start1 = time.process_time()

with open(readpath + filename) as f:
    
    start = time.process_time()
    lines = f.read().strip().split('\n')
    
n_lines = len(lines)
start = time.process_time()
id_ITEMTIMESTEP = [n for n, line in enumerate(lines) if line == 'ITEM: TIMESTEP']

id_header = id_ITEMTIMESTEP[0]+8
header = ['step']+lines[id_header].split()[2:]

id_TIMESTEP = [n + 1 for n in id_ITEMTIMESTEP]
id_datastart = [n + 9 for n in id_ITEMTIMESTEP]
id_dataend = id_ITEMTIMESTEP[1:] + [n_lines]
n_subdata = [i - j for i, j in zip(id_dataend, id_datastart)]
TIMESTEPS = list(chain.from_iterable(repeat(lines[idx], n_subdata[n]) for n, idx in enumerate(id_TIMESTEP) ))
id_data = chain.from_iterable( list(range(start, id_dataend[idx]) )  for idx, start in enumerate(id_datastart) )

data = [[TIMESTEPS[ids]] + lines[idx].split()  for ids, idx in enumerate(id_data)]     #splits on all whitespace if left unspecified  #data = [re.split('\s',lines[idx][:-1]) for idx in id_data]

datapd = pd.DataFrame(data = data, columns = header, dtype = 'float64')
datapd.to_hdf(outputpath + 'custom.h5',key='datapd')
end = time.process_time()
print(end-start)
print(end-start1)
```
