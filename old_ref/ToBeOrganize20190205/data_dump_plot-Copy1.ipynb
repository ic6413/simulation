```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
foldername = 'plot_dump/'
filename = 'nb_maxKEatom.dump' 
f = open(filename)
lines = f.read().strip().split('\n')
lines[-1] = lines[-1] + ' '
idy = 0
while idy < len(lines):
    if lines[idy][0:11]=='ITEM: ATOMS':
        header = lines[idy].split(' ')[2:-1]
        header = ['timestep'] + header
        break
    idy += 1

data = []
idx = 0
idx += 1
while idx < len(lines): 
    timestep = lines[idx]
    while idx < len(lines):
        if lines[idx][0:11]=='ITEM: ATOMS':
            break
        idx += 1
    idx += 1
    while idx < len(lines):
        if lines[idx][0:14]=='ITEM: TIMESTEP':
            break
        data.append( [timestep] + lines[idx][:-1].split(' ') )
        idx += 1
    idx += 1

datapd = pd.DataFrame(data=data, columns=header).astype('float64')
```
